```python
import numpy as np

# Interpolation matrix generation function
def interp_matrix(interp_points, Omegas):
    r = len(Omegas)
    return np.array([[1/np.sqrt(2)] + [func(Omegas[k] * x) for k in range(r) for func in (np.cos, np.sin)] for x in interp_points])

def fourier_basis_matrix(interp_points, omegas):
    N = len(interp_points)  # N = 2M + 1
    rows = [[1/np.sqrt(2)] * N]  # 第一行是常数项

    for omega in omegas:
        rows.append([np.cos(omega * x) for x in interp_points])
        rows.append([np.sin(omega * x) for x in interp_points])

    mat = (2 / N) * np.array(rows)
    return mat

for k in range(1, 11):
    omegas = list(range(1, k + 1))
    interp_points = np.linspace(0, 2 * np.pi, 2 * len(omegas) + 1, endpoint=False) + \
        np.random.rand()
    A = interp_matrix(interp_points, omegas)  # shape: (2r+1, N)
    X = fourier_basis_matrix(interp_points, omegas)  # shape: (2r+1, N)

    error = np.linalg.norm(X @ A - np.eye(2 * len(omegas) + 1))
    print(f"k = {k}, error = {error:.2e}")
```
